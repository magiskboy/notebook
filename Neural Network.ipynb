```python
import numpy as np
import tensorflow as tf
from sklearn.datasets import load_boston
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split
```

```python
data, target = load_boston(return_X_y=True)
data = minmax_scale(data)
target = target.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2)
```

```python
layer_1 = 100
layer_2 = 100

tf.reset_default_graph()

with tf.name_scope('Input'):
    x = tf.placeholder(shape=(None, 13), dtype=tf.float32, name='X')
    y = tf.placeholder(shape=(None, 1), dtype=tf.float32, name='y')
    
with tf.name_scope('Model'):
    w1 = tf.Variable(np.random.rand(13, layer_1), dtype=tf.float32, name='W_1')
    b1 = tf.Variable(np.random.rand(1), dtype=tf.float32, name='b_1')
    w2 = tf.Variable(np.random.rand(layer_1, layer_2), dtype=tf.float32, name='W_2')
    b2 = tf.Variable(np.random.rand(1), dtype=tf.float32, name='b_2')
    w_out = tf.Variable(np.random.rand(layer_2, 1), dtype=tf.float32, name='W_out')
    with tf.name_scope('layer_1'):
        h1 = tf.matmul(x, w1) + b1
        h1 = tf.nn.relu(h1)
    with tf.name_scope('layer_2'):
        h2 = tf.matmul(h1, w2) + b2
        h2 = tf.nn.relu(h2)

with tf.name_scope('Output'):
    output = tf.matmul(h2, w_out)

with tf.name_scope('Optimizer'):
    loss_ops = tf.reduce_mean(tf.squared_difference(y, output))
    optimizer_ops = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(loss_ops)

writer = tf.summary.FileWriter(filename_suffix='boston', graph=tf.get_default_graph(), logdir='logs')
train_summary = tf.summary.scalar('train_loss', loss_ops)
val_summary = tf.summary.scalar('val_loss', loss_ops)
init_ops = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init_ops)
    for step in range(10000):
        sess.run(optimizer_ops, feed_dict={x:x_train, y:y_train})
        loss_train, train_sum = sess.run([loss_ops, train_summary], feed_dict={x:x_train, y:y_train})
        loss_val, val_sum = sess.run([loss_ops, val_summary], feed_dict={x:x_test, y:y_test})       
        writer.add_summary(train_sum)
        writer.add_summary(val_sum)
        print('Train loss {}, val loss {}'.format(loss_train, loss_val))
```
